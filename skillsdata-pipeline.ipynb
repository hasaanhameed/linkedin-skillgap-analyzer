# 3. Cleaning the data

## Dropping columns with NULL skills

```python
skills_df = skills_df.dropna(subset=['job_skills'])

print(f"Missing values in job skills dataset after dropping null job skills:\n {skills_df.isnull().sum()}")
```

## Fixing Iconsistent Text
- Lowercase all records
- Stripping whitespaces

```python
# Job Titles
postings_df['job_title'] = postings_df['job_title'].str.lower().str.strip()
# Job Skills
skills_df['job_skills'] = skills_df['job_skills'].str.lower().str.strip()
```

# 4. Reducing/Filtering the data

## Numerosity Reduction

### Unique job titles and their counts

```python
print(f"Unique job titles in linkedin job postings and their counts:\n {postings_df['job_title'].value_counts()}\n\n")
```

The dataset contains 250,825 unique job titles across all industries. However, for our project we are only interested in tech related job titles with sufficient postings to reliably analyze skill requirements.

### Filtering out all non-technical job titles

#### Creating list of technical job titles

```python
tech_keywords = [
    # Software
    'software engineer', 'software developer', 'software architect', 'software designer', 
    
    # Web
    'frontend', 'front-end', 'backend', 'back-end', 'full stack', 'fullstack', 'web developer', 'web engineer', 'web designer', 
    
    # Data
    'data scientist', 'data analyst', 'data engineer', 'data architect', 'data visualization', 'data science', 'data analytics',
    
    # ML/AI
    'machine learning', 'deep learning', 'artificial intelligence', 'ai engineer', 'ml engineer', 'nlp', 'natural language processing', 'computer vision',
    
    # DevOps/Cloud
    'devops', 'cloud engineer', 'site reliability', 'sre', 'platform engineer', 'aws', 'azure', 'gcp', 'google cloud', 'amazon web services',
    
    # Mobile
    'android developer', 'ios developer', 'mobile developer', 'mobile engineer', 'mobile app developer', 'mobile application developer',
    
    # Security
    'cybersecurity', 'security engineer', 'penetration tester', 'security analyst', 'information security', 'infosec',
    
    # Database
    'database administrator', 'dba', 'database engineer', 'database developer', 'sql developer', 'sql engineer',
    
    # General
    'programmer', 'software qa', 'quality assurance engineer', 'test engineer', 'automation engineer', 'technical lead', 'tech lead', 'technology lead', 'cto', 'chief technology officer',
    
    # Common abbreviations safe to use
    'ui/ux', 'ux designer', 'ui designer',
    'it support', 'it specialist',
    'network engineer', 'blockchain'
]
```

#### Checking count of technical jobs in 500000 job postings

```python
tech_count = postings_df['job_title'].str.contains(
    '|'.join(tech_keywords), case=False, na=False
).sum()

print(f"Tech related job postings: {tech_count}")
print(f"Out of total: {len(postings_df)}")
```

We have 26683 technical jobs in total 500000 jobs 

#### Checking count of technical jobs in the entire dataset (1.3 Million job postings)

```python
postings_df = pd.read_csv('linkedin_job_postings.csv')
```

```python
tech_count = postings_df['job_title'].str.contains(
    '|'.join(tech_keywords), case=False, na=False
).sum()

print(f"Tech related job postings: {tech_count}")
print(f"Out of total: {len(postings_df)}")
```

So, in the total dataset, we have 72385 technical jobs in total 1.3 million job postings

#### Keeping only technical jobs and filtering out the rest

```python
postings_df = postings_df[postings_df['job_title'].str.contains(
    '|'.join(tech_keywords), case=False, na=False)]
```
