# 6. Data Transformation

Jobs of different fields are spread across many job titles, so, we will be categorizing them as: 
1. Software Engineering

- Senior/Lead/Staff Software Engineer (all variants)

6. IT & Security

- IT Specialist
- Information Technology (IT) Specialist (Security - INFOSEC)

## Categorizing Job titles

```python
def categorize_job_title(title):
    title = title.lower()
    if any(x in title for x in ['software engineer', 'software developer', 'backend engineer']):
        return 'Software Engineering'
    elif any(x in title for x in ['network engineer', 'network automation', 'ddi', 'network operations']):
        return 'Network Engineering'
    elif 'data engineer' in title:
        return 'Data Engineering'
    elif any(x in title for x in ['data scientist', 'data analyst', 'machine learning', 'data architect']):
        return 'Data Science & Analytics'
    elif any(x in title for x in ['devops', 'automation engineer', 'test engineer']):
        return 'DevOps & Automation'
    elif any(x in title for x in ['it specialist', 'information technology', 'infosec']):
        return 'IT & Security'
    else:
        return 'Other'
```

```python
merged_job_df['job_title'] = merged_job_df['job_title'].apply(categorize_job_title)
```
